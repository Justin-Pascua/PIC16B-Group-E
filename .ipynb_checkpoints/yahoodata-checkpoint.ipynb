```python
from yahoo_fin.stock_info import get_data
import requests
import pandas as pd
```

```python
#Example of get_data
#Gets all of the historic data available for Amazon with a weekly interval between the two dates
#Data is returned in a pandas dataframe
amazon_weekly = get_data('amzn',
                         start_date='12/04/2009',
                         end_date='02/18/2025',
                         index_as_date = True,
                         interval='1wk')
amazon_weekly
```

```python
def data_for_multiple_tickers(ticker_list,
                              start_date,
                              end_date,
                              index_as_date = True,
                              interval = '1wk'):
    
    '''
    A function to get historical data for multiple tickers at once.
    
    ticker_list is a list of tickers (strings)
    start_date & end_date are MM/DD/YYYY strings
    index_as_date bool
    interval can be '1d', '1wk', '1mo'
    
    Returns a dictionary of pandas dataframes for each ticker
    '''
    
    historical_datas = {}
    
    for ticker in ticker_list:
        df = get_data(ticker,
                      start_date,
                      end_date,
                      index_as_date,
                      interval)
        
        historical_datas[ticker] = df
        
    return historical_datas
```

```python
#Example: 
list1 = ['amzn', 'aapl', 'ba']
dict_of_hist_data = data_for_multiple_tickers(ticker_list = list1,
                              start_date='12/04/2009',
                              end_date='12/04/2019')
dict_of_hist_data
```

```python
#We can still get data for a specific ticker by just looking it up in the dictionary
dict_of_hist_data['aapl']
```
